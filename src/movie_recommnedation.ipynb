```python
import pandas as pd 
import numpy as np
```

```python
def get_recommendations(movies_df, film_name, metric):
    film_id = movies_df[movies_df['original_title'] == film_name].index[0]
    metric[film_id, film_id] = -1
    # let's keep the first 10
    rec_ids = np.argsort(metric[film_id, :])[::-1][:10]
    
    return movies_df.loc[rec_ids, ['original_title', 'vote_average', 'genres', 'cast', 'director']].sort_values(by=['vote_average'], ascending=False)
```

```python
movies = pd.read_csv('../movies_filtered.csv', index_col=None)
genre_metric = np.load('../genre_cos_similarity.npy')
overview_metric = np.load('../overview_cos_similarity.npy')
cast_metric = np.load('../cast_cos_similarity.npy')
```

```python
gen_w, ov_w, cast_w = 0.1, 0.1, 0.9
combined_metric = gen_w*genre_metric + ov_w*overview_metric + cast_w*cast_metric
```

```python
film_name='Memento'
get_recommendations(movies, film_name, combined_metric)
```
